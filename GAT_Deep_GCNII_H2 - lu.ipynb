```python
def get_metrics(real_score, predict_score):
    sorted_predict_score = np.array(
        sorted(list(set(np.array(predict_score).flatten()))))
    sorted_predict_score_num = len(sorted_predict_score)
    thresholds = sorted_predict_score[np.int32(
        sorted_predict_score_num*np.arange(1, 1000)/1000)]
    #print("------------------------------------------thresholds{}".format(thresholds.shape))
    thresholds = np.mat(thresholds)
    thresholds_num = thresholds.shape[1]

    predict_score_matrix = np.tile(predict_score, (thresholds_num, 1))
    negative_index = np.where(predict_score_matrix < thresholds.T)
    positive_index = np.where(predict_score_matrix >= thresholds.T)
    predict_score_matrix[negative_index] = 0
    predict_score_matrix[positive_index] = 1
    TP = predict_score_matrix.dot(real_score.T)
    #print("------------------------------------------TP{}".format(TP.shape))
    FP = predict_score_matrix.sum(axis=1)-TP
    FN = real_score.sum()-TP
    TN = len(real_score.T)-TP-FP-FN

    fpr = FP/(FP+TN)
    tpr = TP/(TP+FN)
    ROC_dot_matrix = np.mat(sorted(np.column_stack((fpr, tpr)).tolist())).T
    ROC_dot_matrix.T[0] = [0, 0]
    ROC_dot_matrix = np.c_[ROC_dot_matrix, [1, 1]]
    x_ROC = ROC_dot_matrix[0].T
    y_ROC = ROC_dot_matrix[1].T
    auc = 0.5*(x_ROC[1:]-x_ROC[:-1]).T*(y_ROC[:-1]+y_ROC[1:])

    recall_list = tpr
    precision_list = TP/(TP+FP)
    PR_dot_matrix = np.mat(sorted(np.column_stack(
        (recall_list, precision_list)).tolist())).T
    PR_dot_matrix.T[0] = [0, 1]
    PR_dot_matrix = np.c_[PR_dot_matrix, [1, 0]]
    x_PR = PR_dot_matrix[0].T
    y_PR = PR_dot_matrix[1].T
    aupr = 0.5*(x_PR[1:]-x_PR[:-1]).T*(y_PR[:-1]+y_PR[1:])

    f1_score_list = 2*TP/(len(real_score.T)+TP-TN)
    accuracy_list = (TP+TN)/len(real_score.T)
    specificity_list = TN/(TN+FP)

    max_index = np.argmax(f1_score_list)
    f1_score = f1_score_list[max_index]
    accuracy = accuracy_list[max_index]
    specificity = specificity_list[max_index]
    recall = recall_list[max_index]
    precision = precision_list[max_index]
    return [aupr[0, 0], auc[0, 0], f1_score, accuracy, recall, specificity, precision]
```

```python
class Myloss(nn.Module):
    def __init__(self):
        super(Myloss, self).__init__()

    def forward(self, target, prediction, miRNA_lap, dis_lap, alpha1, alpha2, phi1, phi2):
        
        loss_ls = t.norm((target - prediction), p='fro') ** 2

        miRNA_reg = t.trace(t.mm(t.mm(alpha1.T, miRNA_lap), alpha1))
        dis_reg = t.trace(t.mm(t.mm(alpha2.T, dis_lap), alpha2))
        graph_reg = phi1 * miRNA_reg + phi2 * dis_reg

        #graph_reg = t.mm(miRNA_lap, miRNA_lap.T)

        loss_sum = loss_ls + graph_reg

        return loss_sum.sum()
```

```python
def constructNet(miRNA_dis_matrix):
    miRNA_matrix = np.matrix(
        np.zeros((miRNA_dis_matrix.shape[0], miRNA_dis_matrix.shape[0]), dtype=np.int8))
    dis_matrix = np.matrix(
        np.zeros((miRNA_dis_matrix.shape[1], miRNA_dis_matrix.shape[1]), dtype=np.int8))

    mat1 = np.hstack((miRNA_matrix, miRNA_dis_matrix))
    mat2 = np.hstack((miRNA_dis_matrix.T, dis_matrix))
    adj = np.vstack((mat1, mat2))
    return adj


def constructHNet(miRNA_dis_matrix, miRNA_matrix, dis_matrix):
    mat1 = np.hstack((miRNA_matrix, miRNA_dis_matrix))
    mat2 = np.hstack((miRNA_dis_matrix.T, dis_matrix))
    return np.vstack((mat1, mat2))


def get_edge_index(matrix):
    edge_index = [[], []]
    for i in range(matrix.shape[0]):
        for j in range(matrix.shape[1]):
            if matrix[i, j] != 0:
                edge_index[0].append(i)
                edge_index[1].append(j)
    return t.LongTensor(edge_index)


def laplacian(kernel):
    d1 = sum(kernel)
    D_1 = t.diag(d1)
    #I_M = t.diag(t.ones(len(d1)))
    L_D_1 = D_1 - kernel# + I_M
    D_5 = D_1.rsqrt()
    D_5 = t.where(t.isinf(D_5), t.full_like(D_5, 0), D_5)
    L_D_11 = t.mm(D_5, L_D_1)
    L_D_11 = t.mm(L_D_11, D_5)
    return L_D_11


def normalized_embedding(embeddings):
    [row, col] = embeddings.size()
    ne = t.zeros([row, col])
    for i in range(row):
        ne[i, :] = (embeddings[i, :] - min(embeddings[i, :])) / (max(embeddings[i, :]) - min(embeddings[i, :]))
        #ne[i, :] = (embeddings[i, :] - t.mean(embeddings[i, :])) / (t.std(embeddings[i, :]))
    return ne


def getGipKernel(y, trans, gamma, normalized=False):
    if trans:
        y = y.T
    if normalized:
        y = normalized_embedding(y)
    krnl = t.mm(y, y.T)
    krnl = krnl / t.mean(t.diag(krnl))
    krnl = t.exp(-kernelToDistance(krnl) * gamma)
    #krnl = cosine_kernel(krnl, krnl)
    return krnl


def kernelToDistance(k):
    di = t.diag(k).T
    d = di.repeat(len(k)).reshape(len(k), len(k)).T + di.repeat(len(k)).reshape(len(k), len(k)) - 2 * k
    return d


def cosine_kernel(tensor_1, tensor_2):
    return t.DoubleTensor([t.cosine_similarity(tensor_1[i], tensor_2, dim=-1).tolist() for i in
                           range(tensor_1.shape[0])])


def normalized_kernel(K):
    K = abs(K)
    k = K.flatten().sort()[0]
    min_v = k[t.nonzero(k, as_tuple=False)[0]]
    K[t.where(K == 0)] = min_v
    D = t.diag(K)
    D = D.sqrt()
    S = K / (D * D.T)
    return S
```

```python
class GraphAttentionLayer(nn.Module):
    """
    Simple GAT layer, similar to https://arxiv.org/abs/1710.10903
    """
    def __init__(self, in_features, out_features, dropout, alpha_GAT, concat=True):
        super(GraphAttentionLayer, self).__init__()
        self.dropout = dropout
        self.in_features = in_features
        self.out_features = out_features
        self.alpha_GAT = alpha_GAT
        self.concat = concat

        self.W = nn.Parameter(t.empty(size=(in_features, out_features)))
        nn.init.xavier_uniform_(self.W.data, gain=1.414)
        self.a = nn.Parameter(t.empty(size=(2*out_features, 1)))
        nn.init.xavier_uniform_(self.a.data, gain=1.414)

        self.leakyrelu = nn.LeakyReLU(self.alpha_GAT)

    def forward(self, h, adj):
        Wh = t.mm(h, self.W) # h.shape: (N, in_features), Wh.shape: (N, out_features)
        e = self._prepare_attentional_mechanism_input(Wh)

        zero_vec = -9e15*t.ones_like(e)
        attention = t.where(adj > 0, e, zero_vec)
        attention = F.softmax(attention, dim=1)
        attention = F.dropout(attention, self.dropout, training=self.training)
        h_prime = t.matmul(attention, Wh)

        if self.concat:
            return F.elu(h_prime)
        else:
            return h_prime

    def _prepare_attentional_mechanism_input(self, Wh):
        # Wh.shape (N, out_feature)
        # self.a.shape (2 * out_feature, 1)
        # Wh1&2.shape (N, 1)
        # e.shape (N, N)
        Wh1 = t.matmul(Wh, self.a[:self.out_features, :])
        Wh2 = t.matmul(Wh, self.a[self.out_features:, :])
        # broadcast add
        e = Wh1 + Wh2.T
        return self.leakyrelu(e)

    def __repr__(self):
        return self.__class__.__name__ + ' (' + str(self.in_features) + ' -> ' + str(self.out_features) + ')'

    
class GAT(nn.Module):
    def __init__(self, nfeat, nhid, nclass, dropout, alpha_GAT, nheads):
        """Dense version of GAT."""
        super(GAT, self).__init__()
        self.dropout = dropout

        self.attentions = [GraphAttentionLayer(nfeat, nhid, dropout=dropout, alpha_GAT=alpha_GAT, concat=True) for _ in range(nheads)]
        for i, attention in enumerate(self.attentions):
            self.add_module('attention_{}'.format(i), attention)

        self.out_att = GraphAttentionLayer(nhid * nheads, nclass, dropout=dropout, alpha_GAT=alpha_GAT, concat=False)

    def forward(self, x, adj):
        x = F.dropout(x, self.dropout, training=self.training)
        x = t.cat([att(x, adj) for att in self.attentions], dim=1)
        x = F.dropout(x, self.dropout, training=self.training)
        x = F.elu(self.out_att(x, adj))
        return x
    
class GraphConvolution(nn.Module):

    def __init__(self, nfeat, in_features, out_features, dropout, alpha_GAT, nheads, residual = True, variant = False):
        super(GraphConvolution, self).__init__() 
        
        #nfeat, nhid, nclass, dropout, alpha_GAT, nheads
        #self.gat =  GAT(nfeat, in_features,  in_features, dropout, alpha_GAT, nheads)
        
        self.variant = variant
        if self.variant:
            self.in_features = 2*in_features 
        else:
            self.in_features = in_features

        self.out_features = out_features
        self.residual = residual
        self.weight = Parameter(t.FloatTensor(self.in_features,self.out_features))
        self.reset_parameters()

    def reset_parameters(self):
        stdv = 1. / math.sqrt(self.out_features)
        self.weight.data.uniform_(-stdv, stdv)

    def forward(self, input, adj , h0 , lamda, alpha, l):
        theta = math.log(lamda/l+1)
        hi = t.spmm(adj, input)
        if self.variant:
            support = t.cat([hi,h0],1)
            r = (1-alpha)*hi+alpha*h0
        else:
            support = (1-alpha)*hi+alpha*h0
            r = support
        output = theta*t.mm(support, self.weight)+(1-theta)*r
        if self.residual:
            output = output+input
         
        #h, adj,h.shape: (N, in_features)
        #output = self.gat(input, adj)    
            
        return output

class GCNII(nn.Module):
    def __init__(self, nfeat, nlayers, nhidden, last_dim, dropout, lamda, alpha, variant, phi1, phi2, miRNA_size, dis_size, gamma, theda,
                miRNA_matrix, dis_matrix, alpha_GAT, nheads):
        super(GCNII, self).__init__()

        self.gat1 =  GAT(nfeat, miRNA_size + dis_size,  512, dropout, alpha_GAT, nheads)
        self.gat2 =  GAT(512, 512,  nhidden, dropout, alpha_GAT, 2)


        self.convs = nn.ModuleList()
        for _ in range(nlayers):
            #nfeat, in_features, out_features, dropout, alpha_GAT, nheads, residual = True, variant = False
            self.convs.append(GraphConvolution(nfeat, nhidden, nhidden,dropout, alpha_GAT, nheads, variant = variant))
        self.fcs = nn.ModuleList()
        self.fcs.append(nn.Linear(nhidden, nhidden))
        self.fcs.append(nn.Linear(nhidden, last_dim))
        self.gat = nn.ModuleList()
        self.gat.append(self.gat1)
        self.gat.append(self.gat2)

        self.params1 = list(self.convs.parameters())
        self.params2 = list(self.fcs.parameters())
        self.params3 = list(self.gat.parameters())
        self.act_fn = nn.ReLU()
        self.dropout = dropout
        self.alpha = alpha
        self.lamda = lamda
        self.nlayers = nlayers
        

        self.phi1 = phi1
        self.phi2 = phi2
   
        self.miRNA_size = miRNA_size
        self.dis_size = dis_size
        self.gamma = gamma
        self.theda = theda

        self.miRNA_l = []
        self.dis_l = []

        self.miRNA_k = []
        self.dis_k = []

        self.alpha1 = t.randn(self.miRNA_size, self.dis_size).double()
        self.alpha2 = t.randn(self.dis_size, self.miRNA_size).double()
        
        self.miRNA_matrix, self.dis_matrix = t.DoubleTensor(miRNA_matrix), t.DoubleTensor(dis_matrix)
        
        self.gcn_1 = GCNConv(last_dim, 128)
        self.gcn_2 = GCNConv(128, 64)
        self.gcn_3 = GCNConv(64, 32)
        
        self.h1_gamma = 2 ** (-5)
        self.h2_gamma = 2 ** (-3)
        self.h3_gamma = 2 ** (-4)

    def forward(self, x, adj_F):
        
        adj = adj_F['Adj']['data']
        adj_edge = adj_F['Adj']['edge_index']

        _layers = []
        miRNA_kernels = []
        dis_kernels = []
        
        miRNA_kernels.append(self.miRNA_matrix)
        dis_kernels.append(self.dis_matrix)

        #h, adj,h.shape: (N, in_features)
        x = self.gat1(x, adj)
        x = self.gat2(x, adj)
        #print("----------------------------x.shape{}".format(x.shape))
        miRNA_kernels.append(t.DoubleTensor(getGipKernel(x[:self.miRNA_size].clone(), 0, self.h1_gamma, True).double()))
        dis_kernels.append(t.DoubleTensor(getGipKernel(x[self.miRNA_size:].clone(), 0, self.h1_gamma, True).double()))

        #x = F.dropout(x, self.dropout, training=self.training)
        #layer_inner = self.act_fn(self.fcs[0](x))
        layer_inner = self.act_fn(x)

        #print("----------------------------layer_inner.shape{}".format(layer_inner.shape))

        _layers.append(layer_inner)
        for i,con in enumerate(self.convs):
            layer_inner = F.dropout(layer_inner, self.dropout, training=self.training)
            
            layer_inner = self.act_fn(con(layer_inner,adj,_layers[0],self.lamda,self.alpha,i+1))
            
            miRNA_kernels.append(t.DoubleTensor(getGipKernel(layer_inner[:self.miRNA_size].clone(), 0, self.gamma, True).double()))
            dis_kernels.append(t.DoubleTensor(getGipKernel(layer_inner[self.miRNA_size:].clone(), 0, self.gamma, True).double()))  
            
        layer_inner = F.dropout(layer_inner, self.dropout, training=self.training)
        layer_inner = self.fcs[-1](layer_inner)

        '''
        H1 = t.relu(self.gcn_1(layer_inner, adj_edge, adj[adj_edge[0], adj_edge[1]]))
        miRNA_kernels.append(t.DoubleTensor(getGipKernel(H1[:self.miRNA_size].clone(), 0, self.h1_gamma, True).double()))
        dis_kernels.append(t.DoubleTensor(getGipKernel(H1[self.miRNA_size:].clone(), 0, self.h1_gamma, True).double()))

        H2 = t.relu(self.gcn_2(H1, adj_edge, adj[adj_edge[0], adj_edge[1]]))
        miRNA_kernels.append(t.DoubleTensor(getGipKernel(H2[:self.miRNA_size].clone(), 0, self.h2_gamma, True).double()))
        dis_kernels.append(t.DoubleTensor(getGipKernel(H2[self.miRNA_size:].clone(), 0, self.h2_gamma, True).double()))
        
        H3 = t.relu(self.gcn_3(H2, adj_edge, adj[adj_edge[0], adj_edge[1]]))
        miRNA_kernels.append(t.DoubleTensor(getGipKernel(H2[:self.miRNA_size].clone(), 0, self.h3_gamma, True).double()))
        dis_kernels.append(t.DoubleTensor(getGipKernel(H2[self.miRNA_size:].clone(), 0, self.h3_gamma, True).double()))
        '''        
        miRNA_kernels.append(t.DoubleTensor(getGipKernel(layer_inner[:self.miRNA_size].clone(), 0, self.gamma, True).double()))
        dis_kernels.append(t.DoubleTensor(getGipKernel(layer_inner[self.miRNA_size:].clone(), 0, self.gamma, True).double()))

        miRNA_k = sum([(8.0/(self.nlayers + pow(i+1,self.theda))) * miRNA_kernels[i] for i in range(len(miRNA_kernels))])
        self.miRNA_k = normalized_kernel(miRNA_k)
        dis_k = sum([(8.0/(self.nlayers + pow(i+1,self.theda))) * dis_kernels[i] for i in range(len(dis_kernels))])        
        self.dis_k = normalized_kernel(dis_k)
        
        self.miRNA_l = laplacian(miRNA_k)
        self.dis_l = laplacian(dis_k)

        out1 = t.mm(self.miRNA_k, self.alpha1)
        out2 = t.mm(self.dis_k, self.alpha2)

        out = (out1 + out2.T) / 2

        return out
```

```python
def random_index(index_matrix, k_fold):
    association_nam = index_matrix.shape[1]
    random_index = index_matrix.T.tolist()
    random.seed(1)
    random.shuffle(random_index)
    k_folds = k_fold
    CV_size = int(association_nam / k_folds)
    temp = np.array(random_index[:association_nam - association_nam % k_folds]).reshape(k_folds, CV_size, -1).tolist()
    temp[k_folds - 1] = temp[k_folds - 1] + random_index[association_nam - association_nam % k_folds:]
    return temp
```

```python
def crossval_index(miRNA_dis_matrix, k_fold):
    random.seed(1)
    k_fold = k_fold
    pos_index_matrix = np.mat(np.where(miRNA_dis_matrix == 1))
    neg_index_matrix = np.mat(np.where(miRNA_dis_matrix == 0))
    
    print("------------pos_index_matrix_shape{}".format(pos_index_matrix.shape))
    print("------------neg_index_matrix_shape{}".format(neg_index_matrix.shape))
    
    mat1 = neg_index_matrix[0, random.sample(range(neg_index_matrix.shape[1]),pos_index_matrix.shape[1])]
    mat2 = neg_index_matrix[1, random.sample(range(neg_index_matrix.shape[1]),pos_index_matrix.shape[1])]
    neg_index_matrix = np.vstack((mat1, mat2))
    
    print("------------平衡之后pos_index_matrix_shape{}".format(pos_index_matrix.shape))
    print("------------平衡之后neg_index_matrix_shape{}".format(neg_index_matrix.shape))
    
    pos_index = random_index(neg_index_matrix, k_fold)
    neg_index = random_index(pos_index_matrix, k_fold)
    
    print("------------pos_index[0]_len{}".format(len(pos_index[0])))
    print("------------neg_index[0]_len{}".format(len(neg_index[0])))

    
    
    index = [pos_index[i] + neg_index[i] for i in range(k_fold)]
    return index
```

```python
def train(model, train_data, optimizer):
    model.train()
    regression_crit = Myloss()

    def train_epoch():
        model.zero_grad()
        #x, adj
        score = model(train_data['feature'], train_data)
        #target, prediction, miRNA_lap, dis_lap, alpha1, alpha2, phi1, phi2
        loss = regression_crit(train_data['Y_train'], score, model.miRNA_l, model.dis_l, 
                               model.alpha1, model.alpha2, model.phi1, model.phi2)
        
        model.alpha1 = t.mm(
            t.mm((t.mm(model.miRNA_k, model.miRNA_k) + model.phi1 * model.miRNA_l).inverse(), model.miRNA_k),
            2 * train_data['Y_train'] - t.mm(model.alpha2.T, model.dis_k.T)).detach()
        model.alpha2 = t.mm(t.mm((t.mm(model.dis_k, model.dis_k) + model.phi2 * model.dis_l).inverse(), model.dis_k),
                            2 * train_data['Y_train'].T - t.mm(model.alpha1.T, model.miRNA_k.T)).detach()
        
        '''
        
        model.alpha1 = t.mm(
            t.mm((t.mm(model.miRNA_k, model.miRNA_k) + model.phi1 * t.eye(model.miRNA_size)).inverse(), model.miRNA_k),
            2 * train_data['Y_train'] - t.mm(model.alpha2.T, model.dis_k.T)).detach()
        model.alpha2 = t.mm(t.mm((t.mm(model.dis_k, model.dis_k) + model.phi2 * t.eye(model.dis_size)).inverse(), model.dis_k),
                            2 * train_data['Y_train'].T - t.mm(model.alpha1.T, model.miRNA_k.T)).detach()
        '''
        
        loss = loss.requires_grad_()
        loss.backward()
        optimizer.step()
        return loss

    for epoch in range(1, 15 + 1):
        train_reg_loss = train_epoch()
        if epoch%1 == 0:
            print("epoch : %d, loss:%.2f" % (epoch, train_reg_loss.item()))
    pass


def PredictScore(train_miRNA_dis_matrix, miRNA_matrix, dis_matrix, nfeat, nlayers, nhidden, last_dim, dropout, 
                 lamda, alpha, variant, phi1, phi2, miRNA_size, dis_size, gamma, theda):
    np.random.seed(42)
    train_data = {}
    train_data['Y_train'] = t.DoubleTensor(train_miRNA_dis_matrix)
    Heter_adj = constructHNet(train_miRNA_dis_matrix, miRNA_matrix, dis_matrix)
    Heter_adj = t.FloatTensor(Heter_adj)
    Heter_adj_edge_index = get_edge_index(Heter_adj)
    train_data['Adj'] = {'data': Heter_adj, 'edge_index': Heter_adj_edge_index}
    
    #print("------------------------{}".format(train_data['Adj']['edge_index']))

    X = constructNet(train_miRNA_dis_matrix)
    X = t.FloatTensor(X)
    train_data['feature'] = X


    #nfeat, nlayers, nhidden, last_dim, dropout, lamda, alpha, variant, phi1, phi2, miRNA_size, dis_size, gamma, theda,
    #miRNA_matrix, dis_matrix, alpha_GAT, nheads
    model = GCNII(nfeat, nlayers, nhidden, last_dim, dropout, lamda, alpha, variant, phi1, phi2, miRNA_size, dis_size, gamma, theda,
                  miRNA_matrix, dis_matrix, alpha_GAT = 0.01, nheads = 5)
    print(model)
    #for parameters in model.parameters():
        #print(parameters, ':', parameters.size())

    optimizer = optim.Adam([{'params':model.params1, 'weight_decay':0.01},{'params':model.params2, 'weight_decay':0.001},
                            {'params':model.params3, 'weight_decay':0.001},], lr = 0.001)

    train(model, train_data, optimizer)
    # model: x, adj
    return model(train_data['feature'], train_data)
```
